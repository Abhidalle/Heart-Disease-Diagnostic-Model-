```python
# Import necessary libraries for data manipulation and machine learning
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Load the Cleveland Heart Disease dataset OR the downloaded dataset named "data.csv" is present as well 
url = "https://raw.githubusercontent.com/kb22/Heart-Disease-Prediction/master/dataset.csv"
medical_data = pd.read_csv(url)
```

```python
# Separate the clinical biomarkers (features) from the diagnosis (target)
biomarkers = medical_data.drop(columns=['target'])
diagnosis = medical_data['target']

# Create a 20% holdout set for blind testing for preventing the ai to memorize all the patient data
X_train, X_test, y_train, y_test = train_test_split(biomarkers, diagnosis, test_size=0.2, random_state=42)

# Standardize the features so large values (Cholesterol) don't overpower small values (Age)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Initialize and train the Logistic Regression diagnostic model
diagnostic_model = LogisticRegression()
diagnostic_model.fit(X_train_scaled, y_train)

# Evaluate the model's performance on the unseen test data
predictions = diagnostic_model.predict(X_test_scaled)
accuracy = accuracy_score(y_test, predictions)

print(f"Diagnostic Model Accuracy on Holdout Set: {accuracy * 100:.2f}%\n")
```
